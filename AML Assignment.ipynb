## Apriori

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

```python
dataset = pd.read_csv('store_data.csv')
transactions = [] 
dataset
```

```python
dataset.shape
```

```python
for i in range(0, 7500):
    transactions.append([str(dataset.values[i,j]) for j in range(0, 4)])
transactions
```

```python

rules = apriori(transactions = transactions, min_support = 0.002, min_confidence = 0.1, min_lift = 2, min_length = 2, max_length=2)
print(rules)
```

```python
results=list(rules)
```

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))

resinDF = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right      Hand Side', 'Support', 'Confidence', 'Lift'])
```

```python
resinDF
```

```python
resinDF.nlargest(n = 10, columns = 'Lift')
```

## XGBoost

```python
df=pd.read_csv("bank-additional.csv")
df=pd.get_dummies(df,df.columns[df.dtypes == 'object'])
df
```

```python
X = df.iloc[:, 0:20]
y = df.iloc[:, 20]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
import xgboost
from xgboost import XGBClassifier
(XGBClassifier)
```

```python
xgbt = XGBClassifier(max_depth = 2,
             learning_rate = 0.2,
             objective  = "multi:softmax",
             num_class = 2,
             booster = "gbtree",
             n_estimarors = 10,
             random_state = 123)
model=xgbt.fit(X_train, y_train)
model
```

```python
xgbt_pred = xgbt.predict(X_test)
xgbt_pred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, xgbt_pred))
print(xgbt.score(X_train, y_train))
print(xgbt.score(X_test, y_test))
```
